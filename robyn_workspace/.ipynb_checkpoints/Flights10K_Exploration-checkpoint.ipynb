# Flights 10,000K Sub-Sample Data Exploration

```python
flight10k = pd.read_csv('flights10K.csv', index_col=0)
flight10k.head()
```

check the shape and column names, dtypes and empty row status using `.info()`

```python
flight10k.shape
flight10k.info()
```

Dropping Early Non-Relevant Columns

```python
# dropping the full empty column
flight10k.drop(columns='no_name', inplace=True)

# filter cancelled flights out since they will have no delay in arrival
flight10k = flight10k[flight10k['cancelled'] == 0]
flight10k.drop(columns=['cancelled','cancellation_code'], inplace=True) # since we don't have cancelled flights, don't need to know cancellation codes
```

```python
flight10k.describe()
```

Looking at all the columns with 'delay' listed

```python
flight10k[['dep_delay', 'arr_delay', 'carrier_delay', 'weather_delay', 'nas_delay', 'security_delay', 'late_aircraft_delay']].describe()
```

Correlation Matrix

```python
fig, ax = plt.subplots(figsize=(12,8))
sub_sample_corr = flight10k.corr()
sns.heatmap(sub_sample_corr, cmap='coolwarm_r', annot_kws={'size':20})
ax.set_title('SubSample Correlation Matrix', fontsize=14)
plt.show()
```

```python
import seaborn as sns
sns.scatterplot(x='wheels_off', y='dep_time', data=flight10k) # dep_time == wheels_off
```

```python
# droppping airline informational columns or duplicate info columns -- deleting actual not CRS as actual has more empty values than CRS
flight10k.drop(columns=['mkt_carrier','op_carrier_fl_num','dep_time','arr_time','actual_elapsed_time','air_time'], inplace=True)
# flights = 0 or 1 for number of flights, should we filter this out or just delete?
flight10k.drop(columns=['flights'], inplace=True)

# look at correlation plot now
fig, ax = plt.subplots(figsize=(12,8))
sub_sample_corr = flight10k.corr()
sns.heatmap(sub_sample_corr, cmap='coolwarm_r', annot_kws={'size':20})
ax.set_title('SubSample Correlation Matrix', fontsize=14)
plt.show()
```

```python
# function to quickly display columns with remaining null values
def whatsleft(df):
    nan_cols = [col for col in df.columns if df[col].isnull().sum() > 0]
    return df[nan_cols].isnull().sum()

whatsleft(flight10k)
```
